```python
import speech_recognition as sr
import webbrowser as wb
import pyttsx3 
import musicLibrary as ml
import requests as rq

recognizer = sr.Recognizer()
engine = pyttsx3.init()
newsapi = "c81f07a59a1b42c98b94e6365ac713dd"
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)
def speak(text):
    engine.say(text)
    engine.runAndWait()
    

def processCommand(c) :
    print(c)
    if "open google" in c.lower():
        wb.open("www.google.com")
    elif "open youtube" in c.lower():
        wb.open("www.youtube.com")
    elif "open whatsapp" in c.lower():
        wb.open("web.whatsapp.com")
    elif "open anime watch" in c.lower():
        wb.open("www.aniwatchtv.to/home")
    elif c.lower().startswith("play") :
        song = c.lower().split(" ")[1:]
        link = ml.music[song]
        wb.open(link)
    elif "news" in c.lower():
        r = rq.get(f"https://newsapi.org/v2/top-headlines?country=in&apiKey={newsapi}")
        if r.status_code == 200 :
            data = r.json()
            articles = data.get('articles',[])
            for article in articles :
                speak(article['title'])
    
if __name__ == "__main__":
    speak("Initialing hello...")
    while True :
        # Listen for the wake word "Bheem"
        # obtain audio from the microphone
        r = sr.Recognizer()
        
        print("Recognizing...")
        try :
            with sr.Microphone() as source :
                print("Listening...")
                audio = r.listen(source, timeout = 2, phrase_time_limit = 1)
            
            word = r.recognize_google(audio)
            if (word.lower() == "hello"):
                speak("Hello! How can i help you?")
                with sr.Microphone() as source :
                    print("Hello is active ...")
                    audio = r.listen(source, timeout = 3, phrase_time_limit = 1)
                    command = r.recognize_google(audio)
                    processCommand(command)
        except Exception as e:
            print(e)
```
